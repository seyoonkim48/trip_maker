```python
import webbrowser
import os
def showMap(map):
    map.save("map.html")        # 지도가 포함된 html 파일 생성
    filepath = os.getcwd()      # 현재 작업 폴더
    file_url = 'file:///' + filepath + '/map.html'
    webbrowser.open_new_tab(file_url)   # html 파일 오픈
```

```python
# 지도 중심의 좌표를 알 때
loc = (37.54, 127.05)   # 위도, 경도
map = folium.Map(location=loc, zoom_start=16)  # 지도 객체 생성
showMap(map)    # 웹 브라우저에 지도 표시
```

```python
# 지오코딩 함수
geolocator = Nominatim(user_agent="loc_test")

def geocoding(place:str):
    location = geolocator.geocode(place)
    return {
        "address" : location.address,
        "lat_lng" : (location.latitude, location.longitude)
    }
```

```python
place = "남산타워"
n_tower = geocoding(place)
```

```python
map = folium.Map(location=n_tower['lat_lng'], zoom_start=16)  # 지도 객체 생성
# folium.Marker(location=n_tower['lat_lng'],
#                 popup=place,
#                 icon=folium.Icon(color="red",
#                                 icon='thumb-tack',
#                                 prefix='fa')
#                 ).add_to(map)
```

```python
folium.features.CircleMarker(
    location=n_tower["lat_lng"],
    radius=5,
    color="red",
    fill=True,
    fill_color="blue",
    fill_opacity=0.4
)
showMap(map)    # 웹 브라우저에 지도 표시
```
