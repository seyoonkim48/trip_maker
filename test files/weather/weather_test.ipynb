```python
import requests

# 서울 위경도
LAT = 37.5665
LON = 126.9780

# Open-Meteo 과거 날씨 API (Archive API)
url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": LAT,
    "longitude": LON,
    "start_date": "2020-03-20",  # 5년치 3월 말
    "end_date": "2024-03-31",
    "daily": [
        "temperature_2m_max",   # 일 최고 기온
        "temperature_2m_min",   # 일 최저 기온
        "precipitation_sum",    # 강수량
    ],
    "timezone": "Asia/Seoul"
}

response = requests.get(url, params=params)
data = response.json()
print(data)
```

```python
data.get("daily", {})
```

```python
import pandas as pd

df = pd.DataFrame(data)

df
```
